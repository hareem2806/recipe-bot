```python
df["servings"] = df["timing"].str.extract(r"Servings:\s*(\d+)").astype(int)

# Extract Yield (text after "Yield:")
df["yield"] = df["timing"].str.extract(r"Yield:\s*(.*)$")

df.drop('timing',axis=1,inplace=True)

print(df)
```

```python
df.head()
```

```python
df.cuisine_path.unique()
```

```python
# Remove leading/trailing slashes
df["cuisine_path"] = df["cuisine_path"].str.strip("/")

# Split into hierarchy levels
df["cuisine_levels"] = df["cuisine_path"].str.split("/")

# Extract top-level category
df["category"] = df["cuisine_levels"].apply(lambda x: x[0])

# Extract last-level (cleaning out the word "Recipes")
df["subcategory"] = df["cuisine_levels"].apply(
    lambda x: x[-1].replace(" Recipes", "").strip()
)

# Drop helper column
df = df.drop(columns=["cuisine_levels","cuisine_path"])
```

```python
df.head()
```

```python
df.tail()
```

```python
import matplotlib.pyplot as plt

# Count recipes per subcategory
top10 = (
    df.groupby(["category", "subcategory"])["recipe_name"]
      .count()
      .reset_index(name="recipe_count")
      .sort_values("recipe_count", ascending=False)
      .head(10)
)

# Plot
plt.figure(figsize=(12,6))
plt.barh(top10["subcategory"] + " (" + top10["category"] + ")", top10["recipe_count"], color="skyblue")
plt.gca().invert_yaxis()  # Largest on top
plt.xlabel("Number of Recipes")
plt.title("Top 10 Subcategories by Recipe Count")
plt.show()
```

```python
# Count recipes per category
category_counts = (
    df.groupby("category")["recipe_name"]
      .count()
      .reset_index(name="recipe_count")
      .sort_values("recipe_count", ascending=False)
)

# Show top category
print("Most popular category is:", category_counts.iloc[0]["category"],
      "with", category_counts.iloc[0]["recipe_count"], "recipes")

# Plot top 10 categories
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.bar(category_counts["category"][:10], category_counts["recipe_count"][:10], color="coral")
plt.xticks(rotation=45, ha="right")
plt.ylabel("Number of Recipes")
plt.title("Top 10 Categories by Recipe Count")
plt.show()
```
